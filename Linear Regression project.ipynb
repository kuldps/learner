```python
import statsmodels.api as sm
```

```python
int_rate=loans['Interest.Rate']
Loan_amt=loans['Loan.Amount']
FICO=loans['FICO.Score']
```

```python
y=np.matrix(int_rate).transpose()
x1=np.matrix(Loan_amt).transpose()
x2=np.matrix(FICO).transpose()
```

```python
x=np.column_stack([x1 ,x2])
```

```python
xn=sm.add_constant(x)
```

```python
model=sm.OLS(y,xn)
```

```python
model_fit=model.fit()
```

```python
print(model_fit.pvalues)
print(model_fit.rsquared)
```
